```python
# Load a simple pre-trained model (sentiment analysis)
from transformers import pipeline

# This downloads a pre-trained model from Hugging Face
classifier = pipeline("sentiment-analysis")

# Test it on some text
result = classifier("I love learning AI!")
print(result)
```

```python
# Try different examples
examples = [
    "This is amazing!",
    "I'm so frustrated",
    "The weather is okay",
    "Best day ever!"
]

for text in examples:
    result = classifier(text)
    print(f"Text: '{text}'")
    print(f"Result: {result[0]['label']} (confidence: {result[0]['score']:.2%})\n")
```
