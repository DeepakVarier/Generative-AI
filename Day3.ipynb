# Prompt

python
# Bad prompt (vague)

"Explain artificial intelligence to a 10-year-old using simple words and 2 examples"

# Prompt Engineering
The art and science of crafting prompts to get better, more consistent, and more useful results from LLMs.
Key principle: The better your prompt, the better your output.

# Zero-Shot Prompting

When to use: Simple questions, straightforward tasks

Asking the model to do something without examples. You just ask directly.
"What are the top 3 benefits of machine learning?" #Model answers directly based on its training

# Few-Shot Prompting
When to use: Complex tasks, specific formats, when you need consistent output

Classify the sentiment of these reviews:

Review: "This product is amazing!" → Sentiment: Positive
Review: "Terrible quality, very disappointed" → Sentiment: Negative
Review: "It works fine" → Sentiment: Positive

Review: "Worst purchase ever" → Sentiment:" # Model learns the pattern and answers: Negative

# Temperature

Controls randomness in the model's responses:
| Temperature | Behavior | Best For |
|-------------|----------|----------|
| 0 | Deterministic (same output every time) | Factual questions, consistency |
| 0.3–0.5 | Balanced (some variety, mostly coherent) | Most tasks |
| 0.7–1.0 | Creative (more variety, more interesting) | Creative writing, brainstorming |
| 1.5+ | Very random (unpredictable, often odd) | Exploration only, usually too random |


```python
# Zero-Shot Prompting

from transformers import pipeline

# Load a text generation model
generator = pipeline("text-generation", model="distilgpt2")

# Technique 1: Zero-shot (no examples, just ask)
prompt1 = "Summarize machine learning for a 10-year-old:"

result1 = generator(prompt1, max_length=100, temperature=0.7, do_sample=True)
print("ZERO-SHOT EXAMPLE:")
print(f"Prompt: '{prompt1}'")
print(f"Output: {result1[0]['generated_text']}\n")
```

```python
# Few-Shot Prompting
from transformers import pipeline

# Use a classification pipeline (easier for few-shot)
classifier = pipeline("zero-shot-classification")

# Few-shot concept: showing examples of what you want
print("ZERO-SHOT CLASSIFICATION (Few-shot concept):")
print("-" * 60)

# Example reviews to classify
reviews = [
    "This product is absolutely amazing! Best purchase ever!",
    "Terrible quality, broke after one day",
    "It's okay, nothing special",
    "Worst experience of my life",
    "Love it! Highly recommend!"
]

# Classify each review by sentiment
for review in reviews:
    result = classifier(review, candidate_labels=["positive", "negative", "neutral"])

    # Get the top result
    top_label = result['labels'][0]
    top_score = result['scores'][0]

    print(f"Review: '{review}'")
    print(f"→ Sentiment: {top_label.upper()} ({top_score:.1%} confidence)\n")
```

What's happening:

The model classifies text into categories you define
It's "few-shot" because the model learns from the examples you show it
This is more controlled than free-form generation

```python
# Temperature controlled

from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

# Same prompt, different temperatures
prompt = "The future of artificial intelligence"

print("TEMPERATURE COMPARISON:")
print("=" * 70)
print(f"Prompt: '{prompt}'\n")

temperatures = [0.3, 0.7, 1.5]

for temp in temperatures:
    print(f"Temperature: {temp}")
    print("-" * 70)

    # Generate 2 examples at this temperature
    for i in range(2):
        result = generator(
            prompt,
            max_length=50,
            temperature=temp,
            do_sample=True,
            num_return_sequences=1
        )

        generated = result[0]['generated_text']
        print(f"  Version {i+1}: {generated}\n")

    print()
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

print("PROMPT ENGINEERING STRATEGIES:")
print("=" * 70)

# Strategy 1: Be specific
print("\n1. SPECIFIC PROMPT (BETTER):")
prompt_good = "Write a short technical explanation of how neural networks use backpropagation:"
result = generator(prompt_good, max_length=80, temperature=0.5, do_sample=True)
print(f"Prompt: {prompt_good}")
print(f"Output: {result[0]['generated_text']}\n")

# Strategy 2: Give context/role
print("2. ROLE-BASED PROMPT:")
prompt_role = "As a machine learning expert, explain why overfitting is a problem:"
result = generator(prompt_role, max_length=80, temperature=0.5, do_sample=True)
print(f"Prompt: {prompt_role}")
print(f"Output: {result[0]['generated_text']}\n")

# Strategy 3: Request specific format
print("3. FORMAT-SPECIFIED PROMPT:")
prompt_format = """List 3 advantages of using Python for machine learning:
1. """
result = generator(prompt_format, max_length=60, temperature=0.5, do_sample=True)
print(f"Prompt: {prompt_format}")
print(f"Output: {result[0]['generated_text']}\n")

# Strategy 4: Step-by-step thinking
print("4. STEP-BY-STEP PROMPT:")
prompt_steps = "Explain how to train a neural network. Start with: Step 1: Prepare the data."
result = generator(prompt_steps, max_length=80, temperature=0.5, do_sample=True)
print(f"Prompt: {prompt_steps}")
print(f"Output: {result[0]['generated_text']}\n")
```

# Prompt Engineering Best Practices
DOs:

✅ Be specific and clear about what you want

✅ Provide context if needed

✅ Show examples (few-shot) for complex tasks

✅ Request a specific format (list, paragraph, code, etc.)

✅ Use step-by-step thinking for complex problems

✅ Adjust temperature based on task (0.3 for facts, 0.7+ for creativity)



DON'Ts:

❌ Be vague: "Tell me about AI" (too broad)

❌ Ask for impossible things

❌ Use high temperature for factual questions

❌ Forget to say what you want explicitly

| Task                     | Temperature | Prompting Style                         |
|--------------------------|-------------|------------------------------------------|
| Answering factual questions | 0.3         | Zero-shot, be specific                  |
| Writing code             | 0.5         | Few-shot with examples                  |
| Creative writing         | 0.8–1.0     | Zero-shot, descriptive                  |
| Classification           | 0.5         | Zero-shot with categories               |
| Summarization            | 0.5         | Zero-shot, request length               |
| Brainstorming ideas      | 0.9+        | Zero-shot, open-ended                   |
